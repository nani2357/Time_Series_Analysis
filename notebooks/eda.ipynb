```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_datareader as pdr
import yfinance as yf
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore")
```

```python
df_tesla = yf.download('TSLA')
df_tesla.head()
```

```python
df_tesla.tail()
```

```python
df_tesla.info()
```

```python
#plotting

fig = go.Figure(data=[go.Scatter(x=df_tesla.index, y = df_tesla['High'], mode ='lines', name='High')])

fig.update_layout(title = 'Tesla Stock High Prices',
                  xaxis_title ='Date',
                  yaxis_title = 'High Prices')

fig.show()
```

```python
#xlimt and ylimt

fig = go.Figure(data=[go.Scatter(x=df_tesla.index, y = df_tesla['High'], mode ='lines', name='High')])

fig.update_layout(title = 'Tesla Stock High Prices',
                  xaxis_title ='Date',
                  yaxis_title = 'High Prices',
                  xaxis = dict(range=['2022-01-01', '2023-08-25']))

fig.show()
```

```python
#xlimt and ylimt

fig = go.Figure(data=[go.Scatter(x=df_tesla.index, y = df_tesla['High'], mode ='lines', name='High', line=dict(color ='green'))])

fig.update_layout(title = 'Tesla Stock High Prices',
                  xaxis_title ='Date',
                  yaxis_title = 'High Prices',
                  xaxis = dict(range=['2022-01-01', '2023-08-25']))

fig.show()
```

```python
df_tesla.head()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_tesla.index, y=df_tesla['Open'], mode='lines', name='Open', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=df_tesla.index, y=df_tesla['Close'], mode='lines', name='Close', line=dict(color='red')))

fig.update_layout(title='Tesla Open and Close Prices',
                  xaxis = dict(range=['2022-01-01', '2023-08-25']))

fig.show()
```

## Time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
##year end frequency
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
##quaterly start frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
##Business End Frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_tesla.resample(rule='BA').max()
```

```python
df_tesla.resample(rule='BQS').max()
```

```python
##plotting
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

```python
df_tesla['High'].rolling(11).max().head(20)
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30).mean()
```
